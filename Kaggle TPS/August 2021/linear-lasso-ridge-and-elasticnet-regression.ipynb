```python
X = data_train.drop(columns = ['id', 'loss'], axis = 1)
Y = data_train['loss']
X.shape, Y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
X_train_pd = pd.DataFrame(X_train)
X_train_pd.describe()
```

```python
X_test_pd = pd.DataFrame(X_test)
X_test_pd.describe()
```

```python
lr = LinearRegression()
lrmodel = lr.fit(X_train, y_train)
lrmodel.score(X_test, y_test)
```

```python
lr_y_pred = lrmodel.predict(X_test)
print("R2 Score is", r2_score(y_test, lr_y_pred))
print("Mean Absolute Error is", mean_absolute_error(y_test, lr_y_pred))
print("Mean Squared Error is", np.sqrt(mean_squared_error(y_test, lr_y_pred)))
```

```python
la = Lasso(alpha = 1.0)
lamodel = la.fit(X_train, y_train)
lamodel.score(X_test, y_test)
```

```python
la_y_pred = lamodel.predict(X_test)
print("R2 Score is", r2_score(y_test, la_y_pred))
print("Mean Absolute Error is", mean_absolute_error(y_test, la_y_pred))
print("Mean Squared Error is", np.sqrt(mean_squared_error(y_test, la_y_pred)))
```

```python
ri = Ridge(alpha = 1.0)
rimodel = ri.fit(X_train, y_train)
rimodel.score(X_test, y_test)
```

```python
ri_y_pred = rimodel.predict(X_test)
print("R2 Score is", r2_score(y_test, ri_y_pred))
print("Mean Absolute Error is", mean_absolute_error(y_test, ri_y_pred))
print("Mean Squared Error is", np.sqrt(mean_squared_error(y_test, ri_y_pred)))
```

```python
elnet = ElasticNet(alpha = 1.0)
elmodel = elnet.fit(X_train, y_train)
elmodel.score(X_test, y_test)
```

```python
el_y_pred = elmodel.predict(X_test)
print("R2 Score is", r2_score(y_test, el_y_pred))
print("Mean Absolute Error is", mean_absolute_error(y_test, el_y_pred))
print("Mean Squared Error is", np.sqrt(mean_squared_error(y_test, el_y_pred)))
```
